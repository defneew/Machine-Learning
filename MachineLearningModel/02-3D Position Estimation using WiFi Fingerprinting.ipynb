### DATA PREPROCESSING

```python
# 0 olan değerleri NaN ile değiştir
signal_columns = [col for col in data.columns if "_signal" in col]
data[signal_columns] = data[signal_columns].replace(0, np.nan)

# Eksik değerleri sütunun ortalama değeri ile doldur
data[signal_columns] = data[signal_columns].apply(lambda x: x.fillna(x.mean()))
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

# Sadece sayısal sütunları al
data_numeric = data.select_dtypes(include=[float, int])

# Korelasyon matrisi
corr_matrix = data_numeric.corr()

# Sadece sinyal sütunları ile x, y, z arasındaki korelasyonu al
target_columns = ["x", "y", "z"]
signal_columns = [col for col in data_numeric.columns if "_signal" in col]
corr_with_target = corr_matrix[target_columns].loc[signal_columns]

# En yüksek korelasyona sahip 20 sinyal sütununu göster
top_20_signals = corr_with_target.abs().sum(axis=1).sort_values(ascending=False).head(20)
print(top_20_signals)

# Korelasyonları ısı haritasıyla görselleştir
plt.figure(figsize=(10, 8))
sns.heatmap(corr_with_target, annot=False, cmap="coolwarm")
plt.title("WiFi Sinyalleri ile x, y, z Korelasyonu")
plt.show()
```

```python
# En yüksek korelasyona sahip 20 sinyal sütununu seç
top_20_signal_columns = top_20_signals.index.tolist()

# Kaldırılacak sinyal sütunlarını belirle
dropped_signal_columns = [col for col in signal_columns if col not in top_20_signal_columns]

# Sinyal numaralarını al
signal_no = [col.split("_")[0] for col in top_20_signal_columns]

# En yüksek 20 sinyale karşılık gelen BSSID sütunlarını belirle
top_20_bssid_columns = [no + "_bssid" for no in signal_no]

# Kaldırılacak BSSID sütunlarını belirle
dropped_bssid_columns = [col for col in bssid_columns if col not in top_20_bssid_columns]

# Kaldırılacak tüm sütunları birleştir 
dropped_columns = dropped_bssid_columns + dropped_signal_columns + ["index"] + target_columns

# Sütunları veri setinden kaldır
data_cleaned = data.drop(columns=dropped_columns)

# Son hali kontrol et
print(data_cleaned.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding yapılacak (çok kategorili) sütunlar
label_cols = ["max_id (categoric)", "min_id (categoric)","num_id"]
for col in label_cols:
    le = LabelEncoder()
    data_cleaned[col] = le.fit_transform(data[col])

# One-Hot Encoding yapılacak (az kategorili) sütunlar
one_hot_cols = ["max_ch (categoric)", "min_ch (categoric)", "num_ch"]
data_cleaned = pd.get_dummies(data_cleaned, columns=one_hot_cols)

```

### TRAIN MODEL

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

models_params = {
    "RandomForest": {
        "model": RandomForestRegressor(random_state=13),
        "params": {
            "n_estimators": [100, 200],
            "max_depth": [10, 20, None],
            "min_samples_split": [2, 5],
            "min_samples_leaf": [1, 2],
            "max_features": ['sqrt', 'log2']
        }
    },
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {}
    },
    "Ridge": {
        "model": Ridge(),
        "params": {
            "alpha": [0.1, 1.0, 10.0]
        }
    },
    "Lasso": {
        "model": Lasso(max_iter=5000),
        "params": {
            "alpha": [0.001, 0.01, 0.1, 1.0]
        }
    },
    "XGBoost": {
        "model": XGBRegressor(objective='reg:squarederror', random_state=13),
        "params": {
            "n_estimators": [100, 200],
            "max_depth": [3, 6, 10],
            "learning_rate": [0.01, 0.1],
            "subsample": [0.8, 1]
        }
    },
    "LightGBM": {
        "model": LGBMRegressor(random_state=13),
        "params": {
            "n_estimators": [100, 200],
            "max_depth": [-1, 10, 20],
            "learning_rate": [0.01, 0.1],
            "subsample": [0.8, 1]
        }
    },
    "GradientBoosting": {
        "model": GradientBoostingRegressor(random_state=13),
        "params": {
            "n_estimators": [100, 200],
            "learning_rate": [0.01, 0.1],
            "max_depth": [3, 6, 9]
        }
    },
    "SVR": {
        "model": SVR(),
        "params": {
            "C": [0.1, 1, 10],
            "kernel": ['linear', 'rbf'],
            "gamma": ['scale', 'auto']
        }
    },
    "KNN": {
        "model": KNeighborsRegressor(),
        "params": {
            "n_neighbors": [3, 5, 7, 9],
            "weights": ['uniform', 'distance'],
            "metric": ['euclidean', 'manhattan']
        }
    },
    "DecisionTree": {
        "model": DecisionTreeRegressor(random_state=13),
        "params": {
            "max_depth": [10, 20, None],
            "min_samples_split": [2, 5, 10],
            "min_samples_leaf": [1, 2, 5]
        }
    },
    "ExtraTrees": {
        "model": ExtraTreesRegressor(random_state=13),
        "params": {
            "n_estimators": [100, 200],
            "max_depth": [10, 20, None],
            "min_samples_split": [2, 5],
            "min_samples_leaf": [1, 2]
        }
    },
    "AdaBoost": {
        "model": AdaBoostRegressor(random_state=13),
        "params": {
            "n_estimators": [50, 100],
            "learning_rate": [0.5, 1.0]
        }
    }
}
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.metrics import make_scorer

# Öklidyen mesafe fonksiyonu
def euclidean_distance(y_true, y_pred):
    return np.mean(np.sqrt(np.sum((y_true.values - y_pred) ** 2, axis=1)))

# Skorer olarak kullanabilmek için
euclidean_scorer = make_scorer(euclidean_distance, greater_is_better=False)

def train_and_evaluate(X_train, X_test, y_train, y_test):
    results = {}

    for name, mp in models_params.items():
        print(f"Model: {name} - Optimizasyon Başladı...")
        
        #grid search ile dictionaryden gelen her modeli ve parametreleri eğit en iyisini döndür
        wrapped_model = MultiOutputRegressor(mp["model"])
        reg = GridSearchCV(
            estimator=wrapped_model,
            param_grid={f'estimator__{key}': val for key, val in mp["params"].items()},
            cv=3,
            n_jobs=-1,
            verbose=1,
            scoring=euclidean_scorer  # negatif çünkü minimize etmeye çalışıyoruz
        )

        reg.fit(X_train, y_train)
        best_model = reg.best_estimator_
        preds = best_model.predict(X_test)

        dist = euclidean_distance(y_test, preds)
        print(f"{name} - Ortalama Öklidyen Mesafe: {dist:.4f}")
        print(f"En İyi Parametreler: {reg.best_params_}\n")

        results[name] = {
            "mean_euclidean_distance": dist,
            "best_params": reg.best_params_,
            "model": best_model
        }

    return results
```

```python
from sklearn.model_selection import train_test_split       

y = data[target_columns]
#seçilen verileri train test olarak ayır
X_train, X_test, y_train, y_test = train_test_split(data_cleaned, y, test_size=0.3, random_state=13)

# Model eğitimi ve değerlendirme
results = train_and_evaluate(X_train, X_test, y_train, y_test)
```

### MODEL EVALUATION AND VISUALIZATION

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# En iyi modeli seç
best_model_name = min(results, key=lambda k: results[k]["mean_euclidean_distance"])
best_model_info = results[best_model_name]
best_model = best_model_info["model"]

print(f"En iyi model: {best_model_name}")
print(f"Ortalama Öklidyen Mesafe: {best_model_info['mean_euclidean_distance']:.4f}")
print(f"En iyi parametreler:")
for param, val in best_model_info["best_params"].items():
    print(f"  {param}: {val}")

# Tahmin yap
preds = best_model.predict(X_test)

# 3B Görselleştirme (x, y, z)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(y_test.iloc[:, 0], y_test.iloc[:, 1], y_test.iloc[:, 2], c='blue', label='Gerçek', s=30, alpha=0.6)
ax.scatter(preds[:, 0], preds[:, 1], preds[:, 2], c='red', label='Tahmin', s=30, alpha=0.6)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f"Gerçek vs Tahmin (Model: {best_model_name})")
ax.legend()
plt.tight_layout()
plt.show()

# Tahmin hatalarının histogramı
errors = np.sqrt(np.sum((y_test.values - preds) ** 2, axis=1))

plt.figure(figsize=(8, 4))
sns.histplot(errors, bins=30, kde=True, color="purple")
plt.title(f"{best_model_name} Modeli - Tahmin Hataları (Öklidyen Mesafe)")
plt.xlabel("Hata (mesafe)")
plt.ylabel("Frekans")
plt.tight_layout()
plt.show()
```
